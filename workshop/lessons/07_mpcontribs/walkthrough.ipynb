# MPContribs Walkthrough

## explore through MPContribs Portal

https://mpcontribs.org

* API and Portal Docs
* click Screenshot to log into Portal -> Sign-In (Satellite Page)

### Portal Index

https://portal.mpcontribs.org/

* 15 contributed datasets/project & respective landing pages
* each project contains multiple contributions for MP material/composition
* search description/authors for keywords to reduce the list of displayed entries
* browse across datasets via Cards Explorer

### Contribution Cards Explorer

https://portal.mpcontribs.org/explorer/

* search and preview across materials and projects
* title of the card links to the project's landing page
* Details button opens the corresponding detail page
* Authors and description are expandable
* references linked as footnotes
* also shows a subset of its nested/hierarchical data
* card API endpoint provides embeddable code snippets (see API docs and MP Detail Pages)

### Contribution Detail Pages

https://portal.mpcontribs.org/explorer/5ac09faad4f1443de2673c01

* unique identifier for each contribution
* use to access its *Contribution Details Page*
* renders interactive version of full content
* static version of fully functional Jupyter notebook
* components:
    - free-form hierarchical data
    - tabular data (automatic graph generation)
    - crystal structures (use crystal viewer in future)
* navigation bar:
    - jump to component
    - show/hide components
    - download


### Dataset Landing Pages

use `jarvis_nus` and `dtu` as examples
    
* landing page with unique URL: share with community, cite in journals
* default overview table of all contributions with
    - search box
    - grouped headers
    - column sort
    - pagination
    - column manager
    - link to MP Materials Details Pages
    - link to Contribution Details Pages
    - link to download CIF files for contributed structures
* optional interactive overview graphs (Plotly)
